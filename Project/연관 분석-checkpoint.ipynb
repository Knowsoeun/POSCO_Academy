### 데이터처리에 필요한 것들 임포트

```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

### 데이터 불러오기

```python
df_raw1 = pd.read_csv("jeju_sales.csv", parse_dates=['purchase_date','supply_date'])
df_raw1.info()
```

```python
df1 = df_raw1.loc[:,'item_g_l'] #대분류
```

# 참고링크
https://hezzong.tistory.com/23

```python
df1
df1=set(df1) # 중복을 제거하기 위해서 set을 사용
df1
```

```python
enc = TransactionEncoder()
df_raw_enc = enc.fit_transform(X = set(df1)) # 반영이 안 됨.

df_asso = pd.DataFrame(df_raw_enc, columns = enc.columns_)
df_asso.head()
```

```python
v_min_sup = 0.1 # 유의미한 결과치가 아님.
df_freq = apriori(df_asso, min_support = v_min_sup, use_colnames = True)
df_freq.round(3)
```

```python
df2 = df_raw1.loc[:,'item_g_m'] #중분류
set(df2)
```

```python
enc = TransactionEncoder()
df_raw_enc = enc.fit_transform(X = df2)

df_asso = pd.DataFrame(df_raw_enc, columns = enc.columns_)
df_asso.head()
```

```python
v_min_sup = 0.2
df_freq = apriori(df_asso, min_support = v_min_sup, use_colnames = True)
df_freq.round(3)
```

```python
df3 = df_raw1.loc[:,'item_g_s']
set(df3) #소분류
```

```python
enc = TransactionEncoder()
df_raw_enc = enc.fit_transform(X = df3)

df_asso = pd.DataFrame(df_raw_enc, columns = enc.columns_)
df_asso.head()
```

```python
v_min_sup = 0.1
df_freq = apriori(df_asso, min_support = v_min_sup, use_colnames = True)
df_freq.round(3)
```
