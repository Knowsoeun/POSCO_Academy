#### (5)-1. 구매 수 군집화

-10까지는 1의 단위로, 그 이상의 구매 수는 50단위로 나누어 군집화함

```python
def Purchase_cnt_ch(dir): 
    res = '' 
    if dir <= 0: res = "0" 
    elif dir <= 1: res = "1" 
    elif dir <= 2: res = "2" 
    elif dir <= 3: res = "3" 
    elif dir <= 4: res = "4" 
    elif dir <= 5: res = "5" 
    elif dir <= 6: res = "6" 
    elif dir <= 7: res = "7" 
    elif dir <= 8: res = "8" 
    elif dir <= 9: res = "9" 
    elif dir <= 10: res = "10" 
    elif dir <= 50: res = "50" 
    elif dir <= 100: res = "100" 
    elif dir <= 150: res = "150" 
    else: res = "200" 
    return res
```

```python
df_raw1['Purchase_cnt_change'] = df_raw1['purchase_cnt'].apply(Purchase_cnt_ch)
```

```python
df_raw1.head()
```

```python
sns.countplot(x = 'Purchase_cnt_change', data = df_raw1) 
df_raw1_count = df_raw1["Purchase_cnt_change"].value_counts()
df_raw1_count
```

- 구매 수 변수를 군집화하여 시각화

#### (5)-2. 구매 단가 군집화

-21000원까지 3000원 단위로 군집화, 그 이후로 8, 9, 10단계로 쪼갬

```python
def change_uni_cost(unit_cost):
    res = ''
    if unit_cost <= 3000:
        res="1"
    elif unit_cost <= 6000 :
        res="2"
    elif unit_cost <= 9000 :
        res="3"
    elif unit_cost <= 12000 :
        res="4"
    elif unit_cost <= 15000 :
        res="5"
    elif unit_cost <= 18000 :
        res="6"
    elif unit_cost <= 21000 :
        res="7"
    elif unit_cost <= 30000 :
        res="8"
    elif unit_cost <= 71200 :
        res="9"
    else:
        res="10"
    return res
```

```python
df_raw1['Purchase_cost_change'] = df_raw1['unit_cost'].apply(change_uni_cost)
```

```python
df_raw1.head()
```

```python
sns.countplot(x = 'Purchase_cost_change', data = df_raw1) 
df_raw1_count = df_raw1["Purchase_cost_change"].value_counts()
df_raw1_count
```

- 구매 단가 변수를 군집화하여 시각화

#### (5)-3. 구매 총액 군집화

-21000원까지 3000원 단위로 군집화, 그 이후로 8, 9, 10단계로 쪼갬

```python
def change_unit_amt(purchase_amt):
    res = ''
    if purchase_amt <= 3000:
        res="1"
    elif purchase_amt <= 6000 :
        res="2"
    elif purchase_amt <= 9000 :
        res="3"
    elif purchase_amt <= 12000 :
        res="4"
    elif purchase_amt <= 15000 :
        res="5"
    elif purchase_amt <= 18000 :
        res="6"
    elif purchase_amt <= 21000 :
        res="7"
    elif purchase_amt <= 30000 :
        res="8"
    elif purchase_amt <= 71200 :
        res="9"
    else:
        res="10"
    return res
```

```python
df_raw1['Purchase_amt_change'] = df_raw1['purchase_amt'].apply(change_unit_amt)
```

## Step3. 파생변수 생성

### Step3-0.  jeju_member, jeju_sales 데이터 불러오기

### Step3-1. 구매일자를 계절별로 묶음
